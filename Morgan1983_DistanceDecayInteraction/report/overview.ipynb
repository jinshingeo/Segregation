# Morgan (1983) 구현 개요

## 논문 정보

- 제목: A Distance-Decay Based Interaction Index to Measure Residential Segregation
- 저자: Barrie S. Morgan (King's College London)
- 출처: Area, Vol. 15, No. 3, pp. 211-217, 1983

## 핵심 내용

- 기존 P* 지수(상호작용 지수)는 인접 소지역(tract) 내부 접촉 확률만 측정하는 한계가 있음
- 거리감쇠(distance-decay) 함수를 적용하여 도시 전체(city-wide) 규모의 집단 간 / 집단 내 상호작용 확률을 추정하는 PC* 지수를 제안
- PC*는 P*와 동일한 방식으로 표준화 가능하며, 0과 1 사이 값을 가짐

## 주요 수식

### P* 지수 (Bell, 1954)

$$_aP^*_b = \sum_i \left(\frac{a_i}{\sum a_i}\right) \left(\frac{b_i}{t_i}\right)$$

- $a_i$: 소지역 $i$의 집단 A 인구
- $b_i$: 소지역 $i$의 집단 B 인구
- $t_i$: 소지역 $i$의 전체 인구

### PC* 지수 (Morgan, 1983)

- 거리감쇠 가중치를 적용하여 도시 전체 규모의 상호작용 확률 산출
- 두 개인이 만날 확률은 거리에 반비례
- 잠재적 접촉 수는 거리에 따라 증가하지만 실제 접촉 확률은 감소
